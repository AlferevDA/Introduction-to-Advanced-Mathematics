1) Решите систему уравнений:

y = x^2 – 1

exp(x) + x(1 – y) = 1

```python
x = np.linspace(-5, 5, 501)
plt.plot(x, x**2 - 1)
plt.plot(x, (np.exp(x) + x - 1)/x)
plt.xlabel('x')
plt.ylabel('y')
plt.ylim(-2.5, 25)
plt.grid(True)
plt.show()
```

```python
from scipy.optimize import fsolve

def equations(p):
    x,y = p
    return(y - x**2 + 1, y - ((np.exp(x)+x-1)/x))
x1, y1 = fsolve(equations, (-2, 0))
x2, y2 = fsolve(equations, (1.5, 2.5))
x3, y3 = fsolve(equations, (3.5, 12.5))
print(x1, y1)
print(x2, y2)
print(x3, y3)
```
