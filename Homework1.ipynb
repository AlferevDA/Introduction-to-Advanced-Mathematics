Напишите код на Python, реализующий расчет длины вектора, заданного его координатами. (в программе)

1). Формула длины вектора, расположенного на плоскости и выходящего из точки с координатами (0, 0), где "a" и "b" - соответсвующие координаты его конца.

```python
def vector2D(a, b):
    return ((a**2) + (b**2))**(0.5)
```

```python
vector2D(1, 1)
```

2). Формула длины вектора, расположенного в пространстве и выходящего из точки с координатами (0, 0, 0), где "x", "y" и "z" - соответствующие координаты его конца.

```python
def vector3D(x, y, z):
    return ((x**2) + (y**2) + (z**2))**0.5
```

```python
vector3D(1, 1, 1)
```

3). Формула суммы векторов в координатах.

```python
def vector(x1, y1, z1, x2, y2, z2):
    xn = x1 + x2
    yn = y1 + y2
    zn = z1 + z2
    return xn, yn, zn
```

```python
vector(20, 20, 20, 0, 0, -20)
```

1) Решите систему уравнений:

y = x^2 – 1

exp(x) + x(1 – y) = 1

```python
x = np.linspace(-5, 5, 501)
plt.plot(x, x**2 - 1)
plt.plot(x, (np.exp(x) + x - 1)/x)
plt.xlabel('x')
plt.ylabel('y')
plt.ylim(-2.5, 25)
plt.grid(True)
plt.show()
```

```python
from scipy.optimize import fsolve

def equations(p):
    x,y = p
    return(y - x**2 + 1, y - ((np.exp(x)+x-1)/x))
x1, y1 = fsolve(equations, (-2, 0))
x2, y2 = fsolve(equations, (1.5, 2.5))
x3, y3 = fsolve(equations, (3.5, 12.5))
print(x1, y1)
print(x2, y2)
print(x3, y3)
```
